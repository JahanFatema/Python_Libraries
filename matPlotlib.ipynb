```python
y=x**2
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.xlabel('X lebel')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1,)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y )
plt.xlabel('X lebel')
plt.ylabel('Y label')
plt.title('Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.8,0.8])
axes.plot(x,y)

axes1 = fig.add_axes([0.1,0.4,0.4,0.3])
axes1.scatter(x,y)

axes2 = fig.add_axes([0.48, 0.01, 0.3, 0.3])
axes2.plot(y,x)

axes.set_title('Larger Plot')
axes1.set_title('Scatter Plot')
axes2.set_title('y vs x Plot')
```

```python
fig,axes =plt.subplots(nrows=1,ncol=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
pizza_size = np.arange(8,13)
pizza_size
```

```python
pizza_price=np.arange(300,550,50)
pizza_price
```

```python
plt.scatter(pizza_size,pizza_price)
plt.title("Pizza Price Chart")
plt.xlabel("Pizza Size in inch")
plt.ylabel("Pizza Price in Taka")
plt.xlim(6,13)
plt.show()
```

```python
plt.plot(pizza_size,pizza_price,color='red',marker='s',linestyle='--')
plt.title("Pizza Price Chart")
plt.xlabel("Pizza Size in inch")
plt.ylabel("Pizza Price in Taka")
plt.show()
```
